**<span style="color:maroon">Write a conditional selection to pick the wines that had quality were 6 and over, and then which ones were 5 and under.**

```python
Red_Wine['good quality'] = [1 if x >= 6 else 0 for x in Red_Wine['quality']]
Red_Wine['bad quality'] = [2 if x <= 5 else 0 for x in Red_Wine['quality']]
Red_Wine.head(10)
```

**<span style="color:purple">Separate the Good Wine from the Bad Wine based on the good quality and bad quality, and put them into separate data sets**

```python
Good_Wine = Red_Wine[Red_Wine["good quality"] == 1]
Good_Wine_Other_Indicators = Good_Wine.drop(['quality','good quality','bad quality'], axis = 1)
Good_Wine_Other_Indicators.info()
Good_Wine_Other_Indicators.describe()
Good_Wine_Other_Indicators.head(10)
```

```python
Bad_Wine = Red_Wine[Red_Wine["bad quality"] == 2]
Bad_Wine_Other_Indicators = Bad_Wine.drop(['quality','good quality','bad quality'], axis = 1)
Bad_Wine_Other_Indicators.info()
Bad_Wine_Other_Indicators.describe()
Bad_Wine_Other_Indicators.head(10)
```

**<span style="color:maroon">Answer: <span style="color:red">The difference between the two is that both data sets are divided between good quality and bad quality, it is clear that the other values have a impact on why the wine is good or bad.**

**<span style="color:magenta">Print out graphs based on the two seperate data sets, and analysize the differences between the two data sets**

```python
Good_Wine2 = Good_Wine_Other_Indicators
Good_Wine2.hist(figsize=(10,10),bins=50,color="red",ec="black")
plt.show()
```

```python
Bad_Wine2 = Bad_Wine_Other_Indicators
Bad_Wine2.hist(figsize=(10,10),bins=50,color="green",ec="black")
plt.show()
```
